```python
import math

alpha = 1 / math.log(10)

def log10_briggs(x):
    y = x
    
    #raizes cuadradas consecutivas
    for _ in range(5):
        y = round(math.sqrt(y), 6)
    
    r = y - 1
    
    #Aproximaci√≥n lineal 
    log_approx = 32 * alpha * r
    return log_approx

valores = [1,2,3,2**2,10/2,2*3,7,2**3,3**2,10]
#valores = [ 6**9]

print("x | log10_aprox | log10_real | error rela")
print("-"*45) 

for x in valores:
    if x == 1:
        print(f"{x} | 0.000000 | 0.000000| 0")
    else:
        aprox = log10_briggs(x)
        real = math.log10(x)
        error = abs(aprox - real) / abs(real)
        
        print(f"{x} | {aprox:.6f} | {real:.6f} | {error:.6e}")
            
```
