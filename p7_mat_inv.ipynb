```python
import numpy as np

def inv_matrix(A: np.ndarray) -> np.ndarray:
    """
    Inversión de una matriz cuadrada mediante Gauss-Jordan,
    usando la función gauss_jordan del recurso.
    """
    if not isinstance(A, np.ndarray):
        A = np.array(A, dtype=float)

    n = A.shape[0]
    assert A.shape[0] == A.shape[1], "La matriz A debe ser cuadrada."

    A_inv = np.zeros((n, n))

 
    for i in range(n):
        e = np.zeros(n)
        e[i] = 1  

        # matriz aume
        Ab = np.hstack((A, e.reshape(-1, 1)))

        # resol
        xi = gauss_jordan(Ab)

        A_inv[:, i] = xi

    return A_inv
```
