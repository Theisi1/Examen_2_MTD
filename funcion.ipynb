```python
import matplotlib.pyplot as plt
import numpy as np

def fun (x):
    return x**5 + 5.25 * x**4 + 4.125 * x**3 - 9.125 * x**2 -14.625 * x - 5.625


#Grafica de la funcion

x = np.linspace(-4,2,200)
y = fun(x)
plt.axhline(0,color='black',ls='--')
plt.plot(x,y)
plt.ylim(-10,10)
plt.grid()
plt.show()
```

```python
from scipy.optimize import bisect

x_lin = np.linspace(-4,2,200) #intrvalos obtenido de la grafica
raices = []
for i in range(len(x_lin)-1):
    if fun(x_lin[i]) *fun(x_lin[i+1]) < 0:
        raiz = bisect(fun, x_lin[i], x_lin[i+1])
        if round(raiz, 5) not in [round(r, 5) for r in raices]:
            raices.append(raiz)


print("Raices:")
for r in raices:
    print(r)
    
    
#Grfica con las raices encontradas

x = np.linspace(-4,2,200)
y = fun(x)
plt.axhline(0,color='black',ls='--')
plt.plot(x,y)
plt.ylim(-10,10)
for r in raices:
    plt.plot(r, fun(r), 'ro')  
plt.grid()
plt.show()
```
